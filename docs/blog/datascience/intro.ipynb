# Introduction

In this series of blogs, I will cover key elements of data science:

* Get and clean the dataset
* Tell a story via data visualization 
* Discover correlation and causation of selected variables
* Build a machine learning model for predictions
* Deploy the model via cloud service 
* Monetize your model if it is possible 

We will use `python` as our main programming language. We might also use `R`
or `D3.js` when it comes to the data visualization. 

## Project Based Learning 

The series of blogs are based on a single project, which is to answer some very
important questions:

* What are the features of firms from IT industry?
* What are the drivers of innovation for firms from IT industry?
* How can a firm from IT industry manage or organize the innovation?
