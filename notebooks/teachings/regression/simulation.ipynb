```python
import os
import numpy as np
import pandas as pd
```

```python
ienc = pd.read_csv("./data/IENC_Subsets_Data.csv")
ienc.head()
```

```python
ienc.columns
```

```python
np.random.normal(0.05, 0.01)
```

```python
def _simulate_rd_intensity(shareholding_corporations, 
                        log_num_disputes, rd_intensity_2003, coef):
    if shareholding_corporations == 1:
        return coef * log_num_disputes + np.random.normal(0.05, 0.01)
    else:
        return rd_intensity_2003
```

```python
ienc.columns
```

```python
ienc["log_num_disputes"].values
```

```python
ienc['simulated_rd_intensity'] = ienc.apply(
    lambda x: _simulate_rd_intensity(
        x["ownership5"],
        x["log_num_disputes"],
        x["rd_intensity_2003"], 
        0.2),
        axis=1
)
ienc.head()
```

```python
ienc['simulated_rd_intensity2'] = ienc.apply(
    lambda x: _simulate_rd_intensity(
        x["ownership5"],
        x["log_num_disputes"],
        x["rd_intensity_2003"], 
        0.2),
        axis=1
)
ienc.head()
```
